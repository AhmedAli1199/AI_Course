```python
import numpy as np

grid = [['D','W','D'],
        ['D','W','W'],
        ['W','D','D']]
agent_pos_x = 0
agent_pos_y = 0

def get_dry(grid):
    count = 0
    for i in range(3):
        for j in range(3):
            if grid[i][j] == 'D':
                count += 1
    return count

```

```python
total_dry = get_dry(grid)
print(total_dry)
```

```python
from collections import deque
#Task2
class coordinates:
    def __init__(self, x, y):
        self.x = x
        self.y = y
def print_grid2(grid):
    for i in range(5):
        for j in range(5):
            print(grid[i][j], end=" ")
        print()
def is_valid(x, y):
    if x < 0 or x >= 5 or y < 0 or y >= 5:
        return False
    return True

def print_queue(queue):
    for i in queue:
        print(i.x, i.y)
    


grid2 = [['R','.','D','.','#'],
         ['#','.','.','.','D'],
         ['.','.','D','.','.'],
         ['#','.','#','.','D'],
         ['.','#','.','.','D']]

queue = deque()
visited = set()
agent_cord = coordinates(0, 0)
previous_cord = coordinates(0, 0)

queue.append(agent_cord)

while(queue):
    cuur_cord = queue.popleft()
    if (cuur_cord.x, cuur_cord.y) in visited:   
        continue
    visited.add((cuur_cord.x, cuur_cord.y))

    print("Agent is at position: ", cuur_cord.x, cuur_cord.y)
    grid2[cuur_cord.x][cuur_cord.y] = 'R' # Robot is here now
    print_grid2(grid2)

    top = coordinates(cuur_cord.x - 1, cuur_cord.y)
    down = coordinates(cuur_cord.x + 1, cuur_cord.y)
    right = coordinates(cuur_cord.x, cuur_cord.y + 1)
    left = coordinates(cuur_cord.x, cuur_cord.y - 1)

    if is_valid(top.x, top.y) and (grid2[top.x][top.y] == 'D' or grid2[top.x][top.y] == '.'):
        queue.append(top)
    if is_valid(right.x, right.y) and (grid2[right.x][right.y] == 'D' or grid2[right.x][right.y] == '.'):
        queue.append(right)
    if is_valid(down.x, down.y) and (grid2[down.x][down.y] == 'D' or grid2[down.x][down.y] == '.'):
        queue.append(down)
    if is_valid(left.x, left.y) and (grid2[left.x][left.y] == 'D' or grid2[left.x][left.y] == '.'):
        queue.append(left)  

    print_queue(queue)

    
    

    previous_cord = cuur_cord   
    grid2[previous_cord.x][previous_cord.y] = '.'
    

```
