```python
import random
def genetic_algorithm(problem, population_size):
    value = problem[0]
    weights = problem[1]
    total_weight = problem[2]
    # Initialize population
    population = initialize_population(population_size, value)
    print("initial fitnesses:")
    print([fitness(value, weights, chromosome, total_weight) for chromosome in population])

    iterations = 0
    # Get fitness of each chromosome
    while iterations < 100:
        fitnesses = []
        for chromosome in population:
            fitnesses.append(fitness(value, weights, chromosome, total_weight))
        probabilities = get_probabilities(fitnesses)
        new_chromosomes = []
        # Select top 2 chromosomes based on probability
        new_chromosomes.append(get_top_2(probabilities)[0])
        new_chromosomes.append(get_top_2(probabilities)[1])
        index1 = probabilities.index(new_chromosomes[0])
        index2 = probabilities.index(new_chromosomes[1])
        new_chromosomes[0] = population[index1]
        new_chromosomes[1] = population[index2]
        # Crossover
        new_chromosomes = list(crossover(new_chromosomes))
        # Mutate
        new_chromosomes[0] = mutation(new_chromosomes[0], 0.1)
        new_chromosomes[1] = mutation(new_chromosomes[1], 0.1)

        # Replace
        overweight = get_overweight(population, total_weight, value, weights)
        if len(overweight) > 0:
            del population[overweight[0][1]]
            del population[overweight[1][1]]
            population.append(new_chromosomes[0])
            population.append(new_chromosomes[1])
        else:
            # Remove from population based on lowest fitness
            sorted_fitnesses = sorted(fitnesses)
            del population[fitnesses.index(sorted_fitnesses[0])]
            del population[fitnesses.index(sorted_fitnesses[1])]
            population.append(new_chromosomes[0])
            population.append(new_chromosomes[1])

        iterations += 1

    return population


def get_value(chromosome,values):
    n = len(chromosome)
    sum = 0
    for i in range(n):
        if chromosome[i] == "1":
            sum += values[i]
    
    return sum

def get_weight(chromosome,weights):
    n = len(chromosome)
    sum = 0
    for i in range(n):
        if chromosome[i] == "1":
            sum += weights[i]
    
    return sum


def get_values_and_weights(population,given_value,given_weight):
    values = []
    weights = []
    for chromosome in population:
        value = get_value(chromosome,given_value)
        values.append(value)
        weight = get_weight(chromosome,given_weight)
        weights.append(weight)
    
    return (values,weights)


def get_overweight(population,total_weight,value,weight):
    (weights, values) = get_values_and_weights(population,value,weight)
    overweight = []
    for i in weights:
        if i > total_weight:
            overweight.append((i,weights.index(i))) 

    #return indexes of overweight chromosomes
    return overweight



def mutation(chromosome, mut_prob):
    chromosome = list(chromosome)  # Convert string to list
    n = len(chromosome)
    for i in range(n):
        r = random.uniform(0, 1)
        if r < mut_prob:
            if chromosome[i] == '1':
                chromosome[i] = '0'
            else:
                chromosome[i] = '1'
    return ''.join(chromosome)  # Convert list back to string
    

def crossover(chromosomes):
    c1 = chromosomes[0]
    c2 = chromosomes[1]
    n = len(c1)
    ans1 = ""
    ans2 = ""
    for i in range(n):
        if i < n/2:
            ans1 += c1[i]
            ans2 += c2[i]
        else:
            ans1 += c2[i]
            ans2 += c1[i]
    
    return (ans1,ans2)



def get_probabilities(fitnesses):
    probabilities = []
    total = sum(fitnesses)
    for i in range(len(fitnesses)):
        probabilities.append(fitnesses[i]/total)
    
    return probabilities

def get_top_2(probabilities):
    sorted_probabilities = sorted(probabilities)
    return (sorted_probabilities[-1],sorted_probabilities[-2])
    

    

def initialize_population(population_size,value):
    population = []
    n = len(value)
    for i in range(population_size):
        strng = ""
        for j in range(n):
            r = random.randint(0,1)
            strng += str(r)
        population.append(strng)
    
    return population

def fitness(value,weights,chromosome,total_weight):
    n = len(value)
    sum = 0
    for i in range(n):
        if chromosome[i] == "1":
            sum += value[i]
    local_weight = 0
    for i in range(n):
        if chromosome[i] == "1":
            local_weight += weights[i]
    if local_weight > total_weight:
        sum = sum - local_weight*2

    return sum


def main():
    value = [14,23,8,9,17,15]
    weight = [1,3,7,4,5,6]
    max_weight = 10
    population_size = 8

    problem = (value,weight,max_weight)
    
    population = genetic_algorithm(problem,population_size)
    print(population)
    new_values = [fitness(value,weight,chromosome,max_weight) for chromosome in population]
    print(new_values)

if __name__ == "__main__":
    main()
```
